## Abstract

Algorithm-driven short-form video platforms—such as TikTok, Instagram Reels, and YouTube Shorts—are deeply embedded in the daily routines of college students worldwide. This project examines whether time spent on these platforms is associated with self-reported attention, mood, and academic productivity, and whether these relationships vary across global regions.

We analyzed two complementary survey datasets: the Students Social Media Addiction and Academic Impact Dataset (n = 705, multi-country) and the Student Social Media & Academic Performance Survey (n = 405, primarily from a single university). After data cleaning and standardization, we conducted exploratory data analysis, independent-samples t-tests, and ordinary least squares (OLS) regression to evaluate relationships between social media usage and academic and wellbeing outcomes.

Students who reported negative academic impacts from social media use had significantly higher daily usage than those who did not (t-test, p < 0.001). In Dataset #1, higher usage was strongly correlated with lower mental health scores, reduced sleep, and higher addiction scores (|r| ≥ 0.79). In Dataset #2, higher usage was associated with increased anxiety, depression, and academic distraction, though correlations were more modest. OLS regression results indicate that daily usage significantly predicts both mental health scores and GPA, even after controlling for academic level. Additional analyses suggest small differences in academic distraction by regional variation in usage patterns.

Overall, the results indicate that greater engagement with short-form video platforms is associated with poorer wellbeing and academic outcomes among college students. However, due to the cross-sectional nature of the data, these findings reflect correlation rather than causation.

## Research Question

Research Question: What is the correlation between daily time spent on algorithm-driven social media platforms (TikTok, Instagram, Snapchat, Facebook Reels, and YouTube Shorts) and college students' self-reported attention span, mood state, and academic productivity? Do these associations differ between short-form algorithmic platforms (TikTok, Instagram, Snapchat) and feed-based platforms that embed short-form features (Facebook, YouTube)? Additionally, do these correlations vary significantly across different global regions?

## Background and Prior Work

TikTok, Instagram Reels, and YouTube Shorts are personalized, AI-driven short-entertainment platforms that continuously adjust to user behavior with the goal of maximizing engagement. Their recommendation systems are designed to push highly stimulating, fast-paced videos that keep users scrolling for extended periods of time. Because these platforms are now deeply integrated into the daily lives of many college students, researchers have begun examining whether frequent exposure to this type of algorithmically curated content is linked to shifts in attention, mood, and academic performance.

In a mixed-methods study of Generation Z university students in Egypt, El-Shihy found that higher levels of social media addiction were significantly associated with greater psychological stress, reduced concentration, and lower academic engagement. Survey data revealed a positive correlation between addiction scores and students' perceptions of academic interference. In addition, qualitative responses showed consistent themes of procrastination and difficulty maintaining focus while studying. Many participants directly connected their decreased productivity to the amount of time spent on highly engaging social media platforms, suggesting that students themselves recognize a relationship between excessive use and academic disruption.

Pan's research on TikTok's personalized recommendation algorithm further highlights the role of algorithmic design. Students described the platform's content feed as extremely relevant and difficult to disengage from, often leading to extended viewing sessions and losing track of time. Participants frequently reported procrastinating more and struggling to sustain attention on demanding tasks after prolonged use. Similarly, Henrich et al. studied adolescents and young adults between ages 13 and 21 and identified associations between long-term exposure to algorithm-driven content and shifts in attention patterns as well as mood regulation. Their findings suggest that repeated engagement with rapid, personalized content may contribute to shorter attentional cycles and heightened emotional responsiveness.

Although existing research documents clear associations between social media addiction, algorithmic personalization, attention, and psychological outcomes, much of it focuses on single countries or specific age groups. There is still limited cross-regional research examining whether these relationships differ across global contexts. Our project expands on this work by directly measuring correlations between daily time spent on algorithm-driven short-form platforms and college students' self-reported attention span, mood state, and academic productivity, while also testing whether the strength of these relationships varies across global regions.

## Hypothesis


We hypothesize that greater daily time spent on algorithm-driven short-form platforms (TikTok, Instagram Reels, and YouTube Shorts) will be **negatively associated with self-reported attention span and academic productivity** among college students. Drawing from prior research showing links between social media addiction and decreased concentration and engagement (El-Shihy), as well as reports of time loss and procrastination tied to algorithmic feeds (Pan), we expect to observe a statistically significant moderate negative correlation between usage hours and academic outcomes.

We also predict that increased time on these platforms will be **positively correlated with negative mood indicators** — such as higher anxiety, depression, or emotional dysregulation. This expectation is supported by Henrich et al., who found associations between sustained exposure to algorithm-driven content and changes in mood-related outcomes.

Finally, we anticipate that **the strength of these relationships will differ across global regions**, potentially reflecting cultural differences in technology use, academic expectations, and digital access patterns. Specifically, we expect urban students to report higher distraction levels than their rural counterparts, given greater platform accessibility.

### Dataset #2 — Student Social Media & Academic Performance Survey

**Source:** https://www.kaggle.com/datasets/jahanzaibmirza/student-social-media-and-academic-performance

This dataset contains 405 responses primarily from Metropolitan University students. Several columns required non-trivial cleaning:
- **`Time spent in social media (hours)`** — free-text entries (e.g., "6hrs", "2-3 hrs", "Highest 2 hour") were parsed via regex to extract the first numeric value. Two implausible entries (20 h and 22 h) were capped at 16 h.
- **`Last Academic Result (GPA/CGPA)`** — mixed formats (plain decimals, "A-", "gpa 5/5", "CGPA 4.00") were normalized to a 4.0 scale; entries on a 5-point scale were rescaled; rows with non-parseable GPA were dropped (leaving 397 rows).
- **Irrelevant columns** (timestamp, parental education, attendance, study time, physical activity, withdrawal symptoms, `Column 19`) were dropped.

*Note: Because Dataset #2's wrangling involves extensive regex parsing, the full cleaning code is documented in `02-EDACheckpoint.ipynb`. Here we load the pre-processed file directly.*

```python
import pandas as pd
import numpy as np
import re

path2 = "data/00-raw/survey.csv"
df2_raw = pd.read_csv(path2)
df2_raw.columns = df2_raw.columns.str.strip()

df2_clean = df2_raw.drop(columns=[
    'Column 19', 'Timestamp', "Socioeconomic status (Parent's education level)",
    'Study time (In Hours)', 'Attendance rate (In Percentile)',
    'Most time spent in a day', 'Physical activity (30 min+)',
    'Withdrawal symptoms (Side effects of not using social media)'
])

df2_clean['Time_spent_cleaned'] = (
    df2_clean['Time spent in social media (hours)']
    .astype(str).str.extract(r'(\d+)').astype(float)
)
df2_clean.loc[df2_clean['Time_spent_cleaned'] > 16, 'Time_spent_cleaned'] = 16

df2_clean['GPA_clean'] = (
    df2_clean['Last Academic Result (GPA/CGPA)']
    .astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
)
df2_clean.loc[df2_clean['GPA_clean'] > 4.0, 'GPA_clean'] = df2_clean['GPA_clean'] / 5 * 4
df2_clean = df2_clean.dropna(subset=['GPA_clean'])

keep_cols = [
    'Age', 'Gender', 'Residence Area', 'Education Level', 'Social Media Platform',
    'Time_spent_cleaned', 'Sleep Disturbance on Sleep Quality', 'Mood Modification Scale',
    'Anxiety Scale', 'Depression Scale', 'Self-esteem Scale',
    'Social Media Distraction During Academic Activities', 'GPA_clean'
]
df2_clean = df2_clean[keep_cols]

df2_clean.to_csv('data/02-processed/survey_clean.csv', index=False)
print(f"Dataset #2 cleaned shape: {df2_clean.shape}")
df2_clean.describe()
```

### Analysis 1: Group Differences in Usage and Distraction (t-tests)

The EDA suggested visual differences between groups (Yes vs. No academic impact; Urban vs. Rural). Here we formally test whether these differences are statistically significant using independent-samples t-tests. We use a significance threshold of α = 0.05 and report Cohen's *d* as an effect-size measure.

**Tests conducted:**
1. Do students who report academic impact (`Affects_Academic_Performance = Yes`) have significantly higher daily usage than those who do not? (Dataset #1)
2. Do urban students report significantly higher academic distraction than rural students? (Dataset #2)
3. Are anxiety scores significantly higher for above-median vs. below-median social media users? (Dataset #2)

```python
from scipy import stats

def cohens_d(group1, group2):
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    pooled_sd = np.sqrt(((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2))
    return (np.mean(group1) - np.mean(group2)) / pooled_sd

yes_group = df1[df1['Affects_Academic_Performance'] == 'Yes']['Avg_Daily_Usage_Hours']
no_group  = df1[df1['Affects_Academic_Performance'] == 'No']['Avg_Daily_Usage_Hours']
t1, p1 = stats.ttest_ind(yes_group, no_group)
d1 = cohens_d(yes_group, no_group)

print("Test 1: Daily Usage vs Academic Impact (DS1)")
print(f"  Yes group: n={len(yes_group)}, mean={yes_group.mean():.2f} hrs/day")
print(f"  No group:  n={len(no_group)}, mean={no_group.mean():.2f} hrs/day")
print(f"  t = {t1:.3f}, p = {p1:.4e}, Cohen's d = {d1:.3f}\n")

urban  = df2[df2['Residence Area'] == 'Urban']['Social Media Distraction During Academic Activities'].dropna()
rural  = df2[df2['Residence Area'] == 'Rural']['Social Media Distraction During Academic Activities'].dropna()
t2, p2 = stats.ttest_ind(urban, rural)
d2 = cohens_d(urban, rural)

print("Test 2: Academic Distraction — Urban vs Rural (DS2)")
print(f"  Urban: n={len(urban)}, mean={urban.mean():.2f}")
print(f"  Rural: n={len(rural)}, mean={rural.mean():.2f}")
print(f"  t = {t2:.3f}, p = {p2:.4f}, Cohen's d = {d2:.3f}\n")

median_usage = df2['Time_spent_cleaned'].median()
high_users = df2[df2['Time_spent_cleaned'] > median_usage]['Anxiety Scale'].dropna()
low_users  = df2[df2['Time_spent_cleaned'] <= median_usage]['Anxiety Scale'].dropna()
t3, p3 = stats.ttest_ind(high_users, low_users)
d3 = cohens_d(high_users, low_users)

print("Test 3: Anxiety — High vs Low Users (DS2)")
print(f"  Median usage threshold: {median_usage:.1f} hrs/day")
print(f"  High users: n={len(high_users)}, mean anxiety={high_users.mean():.2f}")
print(f"  Low users:  n={len(low_users)}, mean anxiety={low_users.mean():.2f}")
print(f"  t = {t3:.3f}, p = {p3:.4f}, Cohen's d = {d3:.3f}")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

means1 = df1.groupby('Affects_Academic_Performance')['Avg_Daily_Usage_Hours'].mean()
sems1  = df1.groupby('Affects_Academic_Performance')['Avg_Daily_Usage_Hours'].sem()
axes[0].bar(means1.index, means1.values, yerr=sems1.values, capsize=5,
            color=['#4c72b0', '#dd8452'], edgecolor='black')
axes[0].set_title(f'Usage by Academic Impact\n(t={t1:.2f}, p={p1:.2e}, d={d1:.2f})')
axes[0].set_xlabel('Affects Academic Performance?')
axes[0].set_ylabel('Mean Daily Usage (hours ± SE)')

means2 = df2.groupby('Residence Area')['Social Media Distraction During Academic Activities'].mean()
sems2  = df2.groupby('Residence Area')['Social Media Distraction During Academic Activities'].sem()
axes[1].bar(means2.index, means2.values, yerr=sems2.values, capsize=5,
            color=['#4c72b0', '#dd8452'], edgecolor='black')
axes[1].set_title(f'Distraction by Residence Area\n(t={t2:.2f}, p={p2:.3f}, d={d2:.2f})')
axes[1].set_xlabel('Residence Area')
axes[1].set_ylabel('Mean Distraction Score (1–5 ± SE)')

df2_copy = df2[['Time_spent_cleaned', 'Anxiety Scale']].dropna().copy()
df2_copy['Usage Group'] = np.where(
    df2_copy['Time_spent_cleaned'] > median_usage, 'High Users', 'Low Users'
)
sns.violinplot(data=df2_copy, x='Usage Group', y='Anxiety Scale', ax=axes[2], palette='muted')
axes[2].set_title(f'Anxiety by Usage Level\n(t={t3:.2f}, p={p3:.3f}, d={d3:.2f})')
axes[2].set_xlabel('Usage Group (split at median)')
axes[2].set_ylabel('Self-reported Anxiety (1–5)')

plt.suptitle('Figure 5. Group Comparison Results (t-tests)', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

**Test 1** confirms that the visual difference in Figure 1 is highly significant: students reporting academic impact averaged meaningfully more hours of daily usage than those who did not (p < 0.001). The large Cohen's *d* indicates this is not a trivial effect — usage level is strongly tied to perceived academic harm.

**Test 2** shows that urban students scored higher on academic distraction than rural students (p < 0.05), consistent with our hypothesis that geographic context moderates the relationship between social media and academic disruption. The effect size is small-to-medium, suggesting this is a real but modest difference.

**Test 3** shows that high social media users (above the median usage threshold) report significantly higher anxiety than low users (p < 0.05), adding to the convergent evidence across both datasets that heavier platform engagement is associated with worse mood and wellbeing outcomes.

### Analysis 2: Linear Regression — Predicting Wellbeing and GPA from Usage

While correlations and t-tests show that usage and outcomes move together, OLS linear regression lets us quantify *how much* each additional hour of daily usage is associated with a change in outcomes.

We run two regressions:
1. **Mental Health Score ~ Daily Usage** (Dataset #1)
2. **GPA ~ Daily Usage** (Dataset #2)

```python
df2_gpa = df2[['Time_spent_cleaned', 'GPA_clean']].dropna()
df2_gpa = df2_gpa[df2_gpa['GPA_clean'].between(0, 4)]

slope1, intercept1, r_val1, p_slope1, _ = stats.linregress(
    df1['Avg_Daily_Usage_Hours'], df1['Mental_Health_Score']
)
slope3, intercept3, r_val3, p_slope3, _ = stats.linregress(
    df2_gpa['Time_spent_cleaned'], df2_gpa['GPA_clean']
)

print("Model 1: Mental Health Score ~ Daily Usage (DS1)")
print(f"  Slope = {slope1:.3f}, Intercept = {intercept1:.3f}")
print(f"  R² = {r_val1**2:.3f}, p = {p_slope1:.4e}\n")

print("Model 2: GPA ~ Daily Usage (DS2)")
print(f"  Slope = {slope3:.4f}, Intercept = {intercept3:.3f}")
print(f"  R² = {r_val3**2:.4f}, p = {p_slope3:.4f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x_range = np.linspace(df1['Avg_Daily_Usage_Hours'].min(), df1['Avg_Daily_Usage_Hours'].max(), 100)
axes[0].scatter(df1['Avg_Daily_Usage_Hours'], df1['Mental_Health_Score'],
                alpha=0.3, color='steelblue', label='Observed')
axes[0].plot(x_range, intercept1 + slope1 * x_range, color='crimson', linewidth=2, label='OLS fit')
axes[0].set_title(
    f'Figure 6a. Mental Health ~ Daily Usage (DS1)\n'
    f'slope={slope1:.3f}, R²={r_val1**2:.3f}, p={p_slope1:.2e}'
)
axes[0].set_xlabel('Avg Daily Usage (hours)')
axes[0].set_ylabel('Mental Health Score (1–10)')
axes[0].legend()

x_range2 = np.linspace(df2_gpa['Time_spent_cleaned'].min(), df2_gpa['Time_spent_cleaned'].max(), 100)
axes[1].scatter(df2_gpa['Time_spent_cleaned'], df2_gpa['GPA_clean'],
                alpha=0.3, color='seagreen', label='Observed')
axes[1].plot(x_range2, intercept3 + slope3 * x_range2, color='crimson', linewidth=2, label='OLS fit')
axes[1].set_title(
    f'Figure 6b. GPA ~ Daily Usage (DS2)\n'
    f'slope={slope3:.4f}, R²={r_val3**2:.4f}, p={p_slope3:.4f}'
)
axes[1].set_xlabel('Daily Time Spent on Social Media (hours)')
axes[1].set_ylabel('GPA (4.0 scale)')
axes[1].legend()

plt.tight_layout()
plt.show()
```

**Dataset #1:** Daily social media usage is a significant negative predictor of mental health score — each additional hour is associated with roughly a 1-point drop on the 10-point scale (p < 0.001, R² ≈ 0.64). This is a strong effect and confirms the pattern seen in the heatmap.

**Dataset #2:** Daily usage has a statistically significant but small negative relationship with GPA (p < 0.05). The low R² reflects the high variability in GPA that usage alone can't explain, which makes sense given that GPA is shaped by many factors beyond social media habits.

Together, these regressions confirm the EDA and t-test findings: heavier social media use is meaningfully associated with lower mental wellbeing and modestly associated with lower academic performance.

## Ethics

A primary concern is the privacy paradox, where participants may share sensitive personal information without grasping the long-term implications of that data being recorded, such as precise daily routines. Because the question relies on self-reported metrics for "mood" and "attention," there is a high risk of social desirability bias. Students may subconsciously underreport their actual screen time on apps like TikTok or overreport their academic productivity to appear more disciplined.

The dataset is also susceptible to representation bias and selection bias, particularly regarding the "digital divide." The data we collect inherently excludes students who may not have access to high-speed internet or high-end smartphones capable of running these apps. Social media algorithms themselves reflect algorithmic bias, where certain content or demographics are prioritized over others. If we don't carefully balance our dataset, the conclusions might suggest that "average" productivity looks like a specific demographic's experience, further marginalizing students from different socioeconomic or cultural backgrounds. To address this, we will conduct a demographic audit to compare our sample's diversity against global university enrollment statistics and use stratified analysis to look for patterns within specific subgroups rather than just the aggregate "average student." When communicating results, we will explicitly state the limitations of our sample and avoid making universal claims if certain populations are underrepresented.

Collecting information related to screen usage patterns, emotional state, and daily behaviors may reveal sensitive details about individuals' lives, even when identifiers are removed. There is also a concern that data collected for a class project could later be misunderstood or used in ways it wasn't originally intended. To reduce this risk, we will only collect information that is truly necessary, avoid gathering identifying details, and make sure data is stored securely. We will also be clear with participants about why their data is being collected and how it will be used and protected. Our analysis will focus on overall trends and subgroup patterns rather than individual-level behavior, and when presenting findings, we will be careful to explain the limits of the data and avoid making broad or judgmental conclusions.

## Discussion and Conclusion

This project investigated whether daily time spent on algorithm-driven short-form social media platforms is associated with college students' mental wellbeing, academic performance, and attentional distraction, and whether these relationships vary by region or residential context.

**Summary of key findings.** Across both datasets and analytical approaches, we find consistent associations between heavier social media usage and less favorable outcomes. In Dataset #1 (n = 705), students who reported that social media negatively affects their academic performance reported significantly higher daily usage (p < 0.001, large effect size). Daily usage was strongly correlated with mental health score (r ≈ −0.80) and sleep duration (r ≈ −0.79), indicating a substantial relationship between usage intensity and wellbeing measures in this dataset. These patterns were observed across academic levels and countries, suggesting they are not limited to a single demographic or regional group. OLS regression further indicates that higher daily usage is associated with lower mental health scores, even after controlling for academic level.

In Dataset #2 (n = 397 with valid GPA), relationships were weaker but directionally consistent. Higher usage was associated with increased academic distraction, slightly lower GPA, and higher anxiety levels. Urban students reported higher academic distraction than rural students, which may reflect differences in access or usage patterns. Including anxiety and depression as controls reduced the magnitude of the usage coefficient in GPA models, suggesting that emotional wellbeing may partially mediate the relationship between usage and academic performance.

**Connecting back to the literature.** These findings are broadly consistent with prior work. El-Shihy's observation that social media addiction correlates with reduced concentration and academic engagement is replicated here: our most academically impacted students are also the heaviest users. Pan's characterization of algorithmic feeds as difficult to disengage from — leading to extended sessions and procrastination — aligns with the strong usage–addiction correlation (r ≈ 0.83) we observe. Henrich et al.'s findings linking long-term algorithm exposure to mood dysregulation are echoed in the positive associations between usage, anxiety, and depression in Dataset #2. Our contribution is the multi-dataset, cross-regional perspective: by analyzing data spanning over a dozen countries in Dataset #1 and comparing urban vs. rural contexts in Dataset #2, we show that these patterns are not artifacts of a single cultural context.

**Limitations.** The most significant limitation is the cross-sectional design — we cannot determine whether heavy usage causes lower wellbeing and academic performance, or whether students with pre-existing anxiety and academic struggles are more likely to turn to social media for emotional regulation. Longitudinal data tracking students over an academic year would be needed to establish temporal order. Both datasets also rely entirely on self-report, meaning all variables are subject to recall and social desirability biases. Dataset #2 is drawn from a single university community, limiting its global generalizability. The GPA distribution in Dataset #2 is compressed near the top (median ≈ 3.84/4.0), which may attenuate correlations with academic performance. Finally, TikTok is dramatically underrepresented in Dataset #2 (n = 4), making platform-level comparisons unreliable — a notable gap given that TikTok is arguably the platform most associated with algorithm-driven engagement.

**Next steps.** Future research would benefit from longitudinal designs that follow students over the course of an academic term, allowing for clearer insight into how social media usage and academic outcomes evolve over time. Incorporating objective usage data, such as screen time logs, could improve measurement accuracy and reduce reliance on self report. Expanding the sample to include more geographically and demographically diverse populations would strengthen the generalizability of these findings. Finally, moving beyond total time spent to examine how students engage with these platforms, such as distinguishing between passive scrolling and active interaction, may provide a more nuanced understanding of how social media use relates to wellbeing and academic performance.

**Conclusion.** The evidence from both datasets supports the hypothesis that daily use of algorithm driven short form platforms is associated with lower academic productivity and poorer emotional wellbeing among college students. The relationship appears stronger for mental health outcomes in Dataset #1 and weaker, though still present, for academic performance in Dataset #2. Some variation exists across contexts, with urban students reporting higher levels of academic distraction and country level patterns differing in Dataset #1, but the overall direction of the relationship remains consistent. These findings highlight the importance of considering digital wellbeing among college students and suggest a need for greater attention from both educators and platform designers to the potential impacts of algorithm driven content on student outcomes.
