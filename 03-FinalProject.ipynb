## Abstract

Algorithm-driven short-form video platforms—such as TikTok, Instagram Reels, and YouTube Shorts—are deeply embedded in the daily routines of college students worldwide. This project examines whether time spent on these platforms is associated with self-reported attention, mood, and academic productivity, and whether these relationships vary across global regions.

We analyzed two complementary survey datasets: the Students Social Media Addiction and Academic Impact Dataset (n = 705, multi-country) and the Student Social Media & Academic Performance Survey (n = 405, primarily from a single university). After cleaning and standardizing the data, we conducted exploratory data analysis, independent-samples t-tests, and ordinary least squares (OLS) regression to examine the relationships between social media usage and academic/ wellbeing outcomes.

Students who reported negative academic impacts from social media use had significantly higher daily usage than those who did not (t-test, p < 0.001). In Dataset #1, higher usage was strongly correlated with lower mental health scores, reduced sleep, and higher addiction scores (|r| ≥ 0.79). In Dataset #2, higher usage was associated with increased anxiety, depression, and academic distraction, though correlations were more limited. OLS regression results indicate that daily usage significantly predicts mental health scores (p < 0.001, R^2 = 0.64), but does not significantly predict GPA (p = 0.18), suggesting that academic performance is likely shaped by many factors, beyond social media habits. Additionally, regional comparisons did not yield statistically significant differences in academic distraction (p = 0.29).

Overall, the results indicate that increased engagement with short-form video platforms is associated with poorer wellbeing and academic outcomes among college students. However, since the data is cross-sectional, our findings reflect correlation, not causation.

## Research Question

Research Question: What is the correlation between daily time spent on algorithm-driven social media platforms (e.g. Tiktok, Instagram, Snapchat, Youtube shorts) and college students' self-reported attention span, mood state, and academic productivity? Do these correlations vary significantly across different global regions?

## Background and Prior Work

Algorithm-driven social media platforms such as TikTok and Instagram Reels play a major role and heavily influence the daily lives of college students, and for many, scrolling through these apps has become a routine part of the day. These platforms rely on personalized algorithms that continuously adapt to user behavior in order to deliver content that maximizes engagement.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Because this design makes social media highly engaging and tailored to individual interests, it has raised concerns as to how constant exposure to short-form, data-driven content affects attention, mood, and academic productivity, especially amongst Gen Z college students.

Prior research suggests that extensive social media use can have meaningful psychological and academic consequences. A mixed-methods study by El-Shihy, which examined Gen Z students in Egypt, found that higher levels of social media addiction were associated with increased stress, reduced concentration, and lower academic engagement.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Many students in the study reported difficulty maintaining focus while studying and challenges with time management due to frequent social media use. These findings indicate that students are often aware of the negative effects of excessive social media use, especially when it interferes with their ability to concentrate on academic tasks.

Other research highlights the specific role of algorithms, rather than social media use in general. In Pan's study on TikTok's personalized algorithm, it was evident that college students reported spending more time on the platform because the content felt highly relevant and difficult to disengage from.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) When asked about their tangible experiences, many shared that they were losing track of time, procrastinating more, and experienced a reduction in maintaining their focus and attention during tasks that typically require sustained engagement. Similarly, Gamboa et al. examined the long-term effects of algorithm-driven content consumption on youth populations between the ages of 13-21 and found associations related to changesin attention patterns, mood regulation, and psychological perceptions over time.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Together, these studies suggest that algorithm-driven platforms may influence how students allocate attention and manage academic responsibilities.

It is important to note, nonetheless, that some research find that the relationship between social media use and academic outcomes are weak, or inconsistent once variables like pre-existing anxiety and/or socioeconomic status status are controlled. This raises questions of whether social media usage is the dominant factor, or just something that correlates with students' stress. Also, prior work shows that almost all studies about this (including ours) replies on self-reported screen time, which tends to be underestimated compared to a device's records, so findings can only be interpreted as measures of association.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)

Our project builds on existing research that examines how time spent on algorithm-driven social media varies across specific platforms and regions, in order to analyze college students’ self-reported attention, mood, and academic productivity across different regions of the world.

## Hypothesis


We hypothesize that greater daily time spent on algorithm-driven short-form platforms (TikTok, Instagram Reels, and YouTube Shorts) will be **negatively associated with self-reported attention span and academic productivity** among college students. Based on prior research showing links between social media addiction and decreased concentration and engagement (El-Shihy), as well as reports of time loss and procrastination tied to algorithmic feeds (Pan), we expect to observe a statistically significant moderate negative correlation between usage hours and academic outcomes.

We also predict that increased time on these platforms will be **positively correlated with negative mood indicators** — such as higher anxiety, depression, or emotional dysregulation. This expectation is supported by Henrich et al., who found associations between sustained exposure to algorithm-driven content and changes in mood-related outcomes.

Finally, we anticipate that **the strength of these relationships will differ across global regions**, potentially reflecting cultural differences in technology use, academic expectations, and digital access patterns. Specifically, we expect urban students to report higher distraction levels than their rural counterparts, given greater platform accessibility.

### Dataset #2 — Student Social Media & Academic Performance Survey

**Source:** https://www.kaggle.com/datasets/jahanzaibmirza/student-social-media-and-academic-performance

This dataset contains 405 responses primarily from Metropolitan University students. Several columns required non-trivial cleaning:
- **`Time spent in social media (hours)`** — free-text entries (e.g., "6hrs", "2-3 hrs", "Highest 2 hour") were parsed via regex to extract the first numeric value. Two implausible entries (20 h and 22 h) were capped at 16 h.
- **`Last Academic Result (GPA/CGPA)`** — mixed formats (plain decimals, "A-", "gpa 5/5", "CGPA 4.00") were normalized to a 4.0 scale; entries on a 5-point scale were rescaled; rows with non-parseable GPA were dropped (leaving 397 rows).
- **Irrelevant columns** (timestamp, parental education, attendance, study time, physical activity, withdrawal symptoms, `Column 19`) were dropped.

*Note: Because Dataset #2's wrangling involved extensive  parsing, the full cleaning code is documented in `02-EDACheckpoint.ipynb`. We loaded the pre-processed file directly here.*

````python
import pandas as pd
import numpy as np
import re

path2 = "data/00-raw/survey.csv"
df2_raw = pd.read_csv(path2)
df2_raw.columns = df2_raw.columns.str.strip()

df2_clean = df2_raw.drop(columns=[
    'Column 19', 'Timestamp', "Socioeconomic status (Parent's education level)",
    'Study time (In Hours)', 'Attendance rate (In Percentile)',
    'Most time spent in a day', 'Physical activity (30 min+)',
    'Withdrawal symptoms (Side effects of not using social media)'
])

df2_clean['Time_spent_cleaned'] = (
    df2_clean['Time spent in social media (hours)']
    .astype(str).str.extract(r'(\d+)').astype(float)
)
df2_clean.loc[df2_clean['Time_spent_cleaned'] > 16, 'Time_spent_cleaned'] = 16

df2_clean['GPA_clean'] = (
    df2_clean['Last Academic Result (GPA/CGPA)']
    .astype(str).str.extract(r'(\d+\.?\d*)')[0].astype(float)
)

#addressed TA feedback & fixed gpa issue: removed implausible entries by rescaling only values between 4.0-5.5,
#then dropping anything still outside the 0-4.0 range
mask_5pt = (df2_clean['GPA_clean'] > 4.0) & (df2_clean['GPA_clean'] <= 5.5)
df2_clean.loc[mask_5pt, 'GPA_clean'] = df2_clean.loc[mask_5pt, 'GPA_clean'] / 5 * 4

df2_clean = df2_clean[df2_clean['GPA_clean'].between(0, 4.0)]
df2_clean = df2_clean.dropna(subset=['GPA_clean'])

keep_cols = [
    'Age', 'Gender', 'Residence Area', 'Education Level', 'Social Media Platform',
    'Time_spent_cleaned', 'Sleep Disturbance on Sleep Quality', 'Mood Modification Scale',
    'Anxiety Scale', 'Depression Scale', 'Self-esteem Scale',
    'Social Media Distraction During Academic Activities', 'GPA_clean'
]
df2_clean = df2_clean[keep_cols]

df2_clean.to_csv('data/02-processed/survey_clean.csv', index=False)
print(f"Dataset #2 cleaned shape: {df2_clean.shape}")
df2_clean.describe()
````

### Analysis 1: Group Differences in Usage and Distraction (t-tests)

The EDA suggested visual differences between groups (Yes vs. No academic impact; Urban vs. Rural). Now, we formally test whether these differences are statistically significant using independent-samples t-tests. We use a significance threshold of α = 0.05 and report Cohen's *d* as an effect-size measure.

**Tests conducted:**
1. Do students who report academic impact (`Affects_Academic_Performance = Yes`) have significantly higher daily usage than those who do not? (Dataset #1)
2. Do urban students report significantly higher academic distraction than rural students? (Dataset #2)
3. Are anxiety scores significantly higher for above-median vs. below-median social media users? (Dataset #2)

````python
from scipy import stats

"""
Calculate Cohen's d effect size between 2 independent groups.

Parameters:
-----------
group1 : array-like
1st group of values
group2 : array-like
2nd group of values
    
Returns:
--------
float : Cohen's d value (positive = group1 > group2)
"""

def cohens_d(group1, group2):
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    pooled_sd = np.sqrt(((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2))
    return (np.mean(group1) - np.mean(group2)) / pooled_sd

yes_group = df1[df1['Affects_Academic_Performance'] == 'Yes']['Avg_Daily_Usage_Hours']
no_group  = df1[df1['Affects_Academic_Performance'] == 'No']['Avg_Daily_Usage_Hours']
t1, p1 = stats.ttest_ind(yes_group, no_group)
d1 = cohens_d(yes_group, no_group)

#3 Independent-samples t-tests on group differences from EDA
#Report Cohen's d with p-values to also measure effect size
print("Test 1: Daily Usage vs Academic Impact (DS1)")
print(f"  Yes group: n={len(yes_group)}, mean={yes_group.mean():.2f} hrs/day")
print(f"  No group:  n={len(no_group)}, mean={no_group.mean():.2f} hrs/day")
print(f"  t = {t1:.3f}, p = {p1:.4e}, Cohen's d = {d1:.3f}\n")

urban  = df2[df2['Residence Area'] == 'Urban']['Social Media Distraction During Academic Activities'].dropna()
rural  = df2[df2['Residence Area'] == 'Rural']['Social Media Distraction During Academic Activities'].dropna()
t2, p2 = stats.ttest_ind(urban, rural)
d2 = cohens_d(urban, rural)

print("Test 2: Academic Distraction — Urban vs Rural (DS2)")
print(f"  Urban: n={len(urban)}, mean={urban.mean():.2f}")
print(f"  Rural: n={len(rural)}, mean={rural.mean():.2f}")
print(f"  t = {t2:.3f}, p = {p2:.4f}, Cohen's d = {d2:.3f}\n")

median_usage = df2['Time_spent_cleaned'].median()
high_users = df2[df2['Time_spent_cleaned'] > median_usage]['Anxiety Scale'].dropna()
low_users  = df2[df2['Time_spent_cleaned'] <= median_usage]['Anxiety Scale'].dropna()
t3, p3 = stats.ttest_ind(high_users, low_users)
d3 = cohens_d(high_users, low_users)

print("Test 3: Anxiety — High vs Low Users (DS2)")
print(f"  Median usage threshold: {median_usage:.1f} hrs/day")
print(f"  High users: n={len(high_users)}, mean anxiety={high_users.mean():.2f}")
print(f"  Low users:  n={len(low_users)}, mean anxiety={low_users.mean():.2f}")
print(f"  t = {t3:.3f}, p = {p3:.4f}, Cohen's d = {d3:.3f}")
````

````python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

#Visualizing the t-test results with bar charts for comparisons by group
# and violin plot for the high vs low user anxiety breakdown
means1 = df1.groupby('Affects_Academic_Performance')['Avg_Daily_Usage_Hours'].mean()
sems1  = df1.groupby('Affects_Academic_Performance')['Avg_Daily_Usage_Hours'].sem()
axes[0].bar(means1.index, means1.values, yerr=sems1.values, capsize=5,
            color=['#4c72b0', '#dd8452'], edgecolor='black')
axes[0].set_title(f'Usage by Academic Impact\n(t={t1:.2f}, p={p1:.2e}, d={d1:.2f})')
axes[0].set_xlabel('Affects Academic Performance?')
axes[0].set_ylabel('Mean Daily Usage (hours ± SE)')

means2 = df2.groupby('Residence Area')['Social Media Distraction During Academic Activities'].mean()
sems2  = df2.groupby('Residence Area')['Social Media Distraction During Academic Activities'].sem()
axes[1].bar(means2.index, means2.values, yerr=sems2.values, capsize=5,
            color=['#4c72b0', '#dd8452'], edgecolor='black')
axes[1].set_title(f'Distraction by Residence Area\n(t={t2:.2f}, p={p2:.3f}, d={d2:.2f})')
axes[1].set_xlabel('Residence Area')
axes[1].set_ylabel('Mean Distraction Score (1–5 ± SE)')

df2_copy = df2[['Time_spent_cleaned', 'Anxiety Scale']].dropna().copy()
df2_copy['Usage Group'] = np.where(
    df2_copy['Time_spent_cleaned'] > median_usage, 'High Users', 'Low Users'
)
sns.violinplot(data=df2_copy, x='Usage Group', y='Anxiety Scale', ax=axes[2], palette='muted')
axes[2].set_title(f'Anxiety by Usage Level\n(t={t3:.2f}, p={p3:.3f}, d={d3:.2f})')
axes[2].set_xlabel('Usage Group (split at median)')
axes[2].set_ylabel('Self-reported Anxiety (1–5)')

plt.suptitle('Figure 5. Group Comparison Results (t-tests)', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

**Test 1** confirms that the visual difference in Figure 1 is highly significant: students reporting academic impact averaged meaningfully more hours of daily usage than those who did not (p < 0.001). The large Cohen's *d* also indicates this is not negligible — usage level is strongly tied to perceived academic harm.

**Test 2** shows that while urban students scored slightly higher on academic distraction than rural students (mean = 2.75 vs 2.55), the difference was not statistically significant (t = 1.05, p = 0.29, Cohen's d = 0.15). This null suggests that residence area only, doesn't meaningfully predict academic distraction in this sample, and the small effect size indicates that the difference is moderate. 

**Test 3** shows that high social media users (above the median usage threshold) report significantly higher anxiety than low users (p < 0.05), reinforcing convergent findings across both datasets that more platform engagement is associated with worse mood and wellbeing outcomes.

### Analysis 2: Linear Regression — Predicting Wellbeing and GPA from Usage

While correlations and t-tests show that usage and outcomes move together, OLS linear regression lets us quantify *how much* each additional hour of daily usage is associated with a change in outcomes.

We run two regressions:
1. **Mental Health Score ~ Daily Usage** (Dataset #1)
2. **GPA ~ Daily Usage** (Dataset #2)

````python
#OLS regression to quantify usage-outcome relationship
#Slope tells us how each extra hour of usage changes the outcome (association)
df2_gpa = df2[['Time_spent_cleaned', 'GPA_clean']].dropna()
df2_gpa = df2_gpa[df2_gpa['GPA_clean'].between(0, 4)]

slope1, intercept1, r_val1, p_slope1, _ = stats.linregress(
    df1['Avg_Daily_Usage_Hours'], df1['Mental_Health_Score']
)
slope3, intercept3, r_val3, p_slope3, _ = stats.linregress(
    df2_gpa['Time_spent_cleaned'], df2_gpa['GPA_clean']
)

print("Model 1: Mental Health Score ~ Daily Usage (DS1)")
print(f"  Slope = {slope1:.3f}, Intercept = {intercept1:.3f}")
print(f"  R² = {r_val1**2:.3f}, p = {p_slope1:.4e}\n")

print("Model 2: GPA ~ Daily Usage (DS2)")
print(f"  Slope = {slope3:.4f}, Intercept = {intercept3:.3f}")
print(f"  R² = {r_val3**2:.4f}, p = {p_slope3:.4f}")
````

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

#Regression lines allow us to see overall trend and variability
x_range = np.linspace(df1['Avg_Daily_Usage_Hours'].min(), df1['Avg_Daily_Usage_Hours'].max(), 100)
axes[0].scatter(df1['Avg_Daily_Usage_Hours'], df1['Mental_Health_Score'],
                alpha=0.3, color='steelblue', label='Observed')
axes[0].plot(x_range, intercept1 + slope1 * x_range, color='crimson', linewidth=2, label='OLS fit')
axes[0].set_title(
    f'Figure 6a. Mental Health ~ Daily Usage (DS1)\n'
    f'slope={slope1:.3f}, R²={r_val1**2:.3f}, p={p_slope1:.2e}'
)
axes[0].set_xlabel('Avg Daily Usage (hours)')
axes[0].set_ylabel('Mental Health Score (1–10)')
axes[0].legend()

x_range2 = np.linspace(df2_gpa['Time_spent_cleaned'].min(), df2_gpa['Time_spent_cleaned'].max(), 100)
axes[1].scatter(df2_gpa['Time_spent_cleaned'], df2_gpa['GPA_clean'],
                alpha=0.3, color='seagreen', label='Observed')
axes[1].plot(x_range2, intercept3 + slope3 * x_range2, color='crimson', linewidth=2, label='OLS fit')
axes[1].set_title(
    f'Figure 6b. GPA ~ Daily Usage (DS2)\n'
    f'slope={slope3:.4f}, R²={r_val3**2:.4f}, p={p_slope3:.4f}'
)
axes[1].set_xlabel('Daily Time Spent on Social Media (hours)')
axes[1].set_ylabel('GPA (4.0 scale)')
axes[1].legend()

plt.tight_layout()
plt.show()
````

**Dataset #1:** Daily social media usage is a significant negative predictor of mental health score — each additional hour is associated with roughly a 1-point drop on the 10-point scale (p < 0.001, R² ≈ 0.64). This is a strong effect and confirms the pattern seen in the heatmap.

**Dataset #2:** Daily usage does not have a statistically significant relationship with GPA in this dataset (slope = -0.0075, p = 0.18, R^2 = 0.005). Furthermore, the low R² reflects how just social media usage doesn't entirely explain variance in GPA, which makes sense given that academic performance is determined by many interacting factors. The null result also suggess that effects of social media on grades is likely indirect, and can be related to other variables such as disturbed sleep or anxiety.

Together, these regressions confirm the EDA and t-test findings: heavier social media use is meaningfully associated with lower mental wellbeing and somewhat associated with lower academic performance.

## Ethics

A primary concern with this type of research is the privacy paradox, where participants in the original surveys may have shared sensitive personal information without fully comprehending the long-term implications of that data being recorded. Because our analysis relies on self-reported metrics for mood, attention, and screen time, social desirability bias is a real risk — students may have underreported actual usage or overestimated their academic productivity to appear more disciplined. We acknowledged this throughout our analysis and treated self-reported figures as approximations rather than ground truth.

The datasets are also susceptible to representation and selection bias, particularly around the digital divide. Both samples inherently exclude students without reliable access to smartphones or high-speed internet, meaning our findings reflect connected, app-using populations rather than all college students globally. Social media algorithms themselves introduce algorithmic bias — certain demographics are served different content, which could affect usage patterns in ways our data cannot capture. To address this, we conducted stratified analysis where possible and explicitly avoided making universal claims about student populations. Where subgroups were underrepresented — such as rural students (n = 60) or TikTok users (n = 4) in Dataset #2 — we acknowledged these limitations directly rather than treating aggregate results as a representative.

Since both datasets were sourced from Kaggle and collected by third parties, we did not gather any identifying information ourselves. Our analysis focused entirely on aggregate trends and subgroup patterns rather than individual-level behavior. The datasets were used read-only for academic purposes, consistent with their public licensing. When presenting findings, we were careful to frame results as correlational and to avoid judgmental or deterministic language about students' social media habits.

This project investigated whether daily time spent on algorithm-driven short-form social media platforms is associated with college students' mental wellbeing, academic performance, and attentional distraction, and whether these relationships vary by region or residential context.

**Summary of key findings.** Across both datasets and analytical approaches, we find consistent associations between heavier social media usage and less favorable outcomes. In Dataset #1 (n = 705), students who reported that social media negatively affects their academic performance reported significantly higher daily usage (p < 0.001, large effect size). Daily usage was strongly correlated with mental health score (r ≈ −0.80) and sleep duration (r ≈ −0.79), indicating a substantial relationship between usage intensity and wellbeing measures in this dataset. These patterns were observed across academic levels and countries, suggesting they are not limited to a single demographic or regional group. OLS regression further confirms that daily usage is a negative predictor of mental health scores, in Dataset #1 (slope = -0.70, R^2 = 0.64, p < 0.001). 

In contrast, Dataset #2 (n = 397 with valid GPA) showed weaker but directionally consistent relationships. The GPA regression did not reach statistical significance (p = 0.18, R² = 0.005), suggesting that any influence of social media on academic grades is either too small/ indirect for this sample size to detect. Higher usage was associated with increased academic distraction and higher anxiety levels. The urban vs. rural comparison also did not yield a significant difference in academic distraction (p = 0.29), indicating that residence area alone does not meaningfully predict distraction in this dataset — though the rural subsample (n = 60) was considerably smaller, limiting statistical power.

**Connecting back to the literature.** These findings are broadly consistent with prior work. El-Shihy's observation that social media addiction correlates with reduced concentration and academic engagement is replicated here: our most academically impacted students are also the heaviest users. Pan's characterization of algorithmic feeds as difficult to disengage from — leading to extended sessions and procrastination — aligns with the strong usage–addiction correlation (r ≈ 0.83) we observe. Henrich et al.'s findings linking long-term algorithm exposure to mood dysregulation are echoed in the positive associations between usage, anxiety, and depression in Dataset #2. Our contribution is the multi-dataset, cross-regional perspective: by analyzing data spanning over a dozen countries in Dataset #1 and comparing urban vs. rural contexts in Dataset #2, we show that these patterns are not artifacts of a single cultural context.

**Limitations.** The most significant limitation is the cross-sectional design — we cannot determine whether heavy usage causes lower wellbeing and academic performance, or whether students with pre-existing anxiety and academic struggles are more likely to turn to social media for emotional regulation. Longitudinal data tracking students over an academic year would be needed to establish temporal order. Both datasets also rely entirely on self-report, meaning all variables are subject to recall and social desirability biases. Dataset #2 is drawn from a single university community, limiting its global generalizability. The GPA distribution in Dataset #2 is compressed near the top (median ≈ 3.84/4.0), which may attenuate correlations with academic performance. Finally, TikTok is dramatically underrepresented in Dataset #2 (n = 4), making platform-level comparisons unreliable — a notable gap given that TikTok is arguably the platform most associated with algorithm-driven engagement.

**Next steps.** Future research would benefit from longitudinal designs that follow students over the course of an academic term, allowing for clearer insight into how social media usage and academic outcomes evolve over time. Incorporating objective usage data, such as screen time logs, could improve measurement accuracy and reduce reliance on self report. Expanding the sample to include more geographically and demographically diverse populations would strengthen the generalizability of these findings. Finally, moving beyond total time spent to examine how students engage with these platforms, such as distinguishing between passive scrolling and active interaction, may provide a more nuanced understanding of how social media use relates to wellbeing and academic performance.

The relationship was strongest for mental health in Dataset #1, where usage closely tracked lower wellbeing scores and reduced sleep. In Dataset #2, effects were more moderate — the GPA regression didn't reach significance, which isn't entirely surprising given how many factors shape academic performance beyond screen time. The directional patterns still held for distraction and anxiety, suggesting the relationship is real even if it's hard to isolate in a GPA number. We're cautious about overstating these findings given the self-reported, cross-sectional nature of both datasets, but the consistency across two independent samples lends some confidence to the overall pattern. Algorithm-driven platforms are designed to be hard to put down, and our findings suggest that this has real costs for the students using them most — something worth taking seriously by universities, researchers, and platform designers alike.
